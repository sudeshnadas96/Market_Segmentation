```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the dataset (adjust the path as necessary)
df = pd.read_csv('EV_Charging_Station.csv')
df.head()
# Extract numeric values from the 'power' column and convert to float
df['power'] = df['power'].str.extract(r'(\d+\.?\d*)').astype(float)

# Check for missing values
print(df.isnull().sum())
```

```python
# Handle missing values
df.dropna(subset=['latitude', 'longitude', 'power'], inplace=True)
df.fillna(df.median(numeric_only=True), inplace=True) 
```

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Check for infinite values in numeric columns
print(np.isinf(df[numeric_cols]).sum())

# Replace infinite values with NaN in numeric columns and handle them
df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
df.fillna(df.median(numeric_only=True), inplace=True)  # Fill remaining missing values with median of numeric columns
```

```python
# Check for extremely large/small values
print(df[numeric_cols].describe())

# Clipping values to a reasonable range if necessary
# Example: Clip power to be within a specific range
df['power'] = np.clip(df['power'], 0, df['power'].quantile(0.99))
```

```python
# Encoding categorical features
df_encoded = pd.get_dummies(df, columns=['type', 'service'], drop_first=True)

# Selecting relevant features
features = ['latitude', 'longitude', 'power'] + [col for col in df_encoded.columns if col.startswith('type_') or col.startswith('service_')]
X = df_encoded[features]
```

```python
# Using RobustScaler
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Principal Component Analysis(PCA) for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the PCA result
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('PCA of EV Charging Stations')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-Means with the chosen number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)

# Add the cluster labels to the original dataframe
df['cluster'] = y_kmeans
```

```python
#Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_kmeans, palette='viridis', s=100)
plt.title('Clusters of EV Charging Stations')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyze the clusters
cluster_summary = df.groupby('cluster').mean()
print(cluster_summary)
```
